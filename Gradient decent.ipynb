```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(x):
    return x ** 2 
```

```python
t = np.linspace(-5,5,100)
plt.plot(t, loss(t))
```

```python
def loss_diff(x):
    return 2 * x
```

```python
plt.plot(t, loss(t))
t1 = 4
for i in range(100):
    t1 = t1 - 0.03 * loss_diff(t1)
    plt.scatter(t1, loss(t1))
    
```
